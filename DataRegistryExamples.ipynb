# Registering a dataset within an *execution*

````python
config_file = "/Users/jzuntz/src/lsst/TXPipe/examples/metadetect/config.yml"

execution_id = dr.register_execution(
    name = "txpipe-dp2-metadetect-v1",
    description = "First run of TXPipe on metadetect cats for DP2",
    execution_start = datetime.fromisoformat("2026-06-03"),
    input_datasets = [1, 2], # ids of previously-registered datasets
    configuration = config_file,
)

# Add a new file to the execution
id3, _ = dr.register_dataset(
    "test_shear_map",
    "1.0.0",
    description="My shear map",
    owner="dp2-shear",
    owner_type="project",
    old_location=example_data + "/" + "source_maps.hdf5",
    execution_id=execution_id,
)
    
````

# Full query syntax

````python
f1 = dr.query.gen_filter("dataset.name", "~=", "*map")
f2 = dr.query.gen_filter("version_string", "=", "1.0.0")
dr.find_datasets(
    property_names=["dataset.dataset_id", "dataset.name"],
    filters=[f1,f2],
    return_format='DataFrame',
)
````
